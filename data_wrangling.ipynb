```python
def categorize_experience(years):
    try:
        years = float(years)  
        if years <= 2:
            return "Beginner"
        elif years <= 5:
            return "Intermediate"
        elif years <= 10:
            return "Advanced"
        else:
            return "Expert"
    except ValueError:
        return None  
```

```python
data['YearsCodePro'] = pd.to_numeric(data['YearsCodePro'], errors='coerce')  # Convert to numeric
data['ExperienceLevel'] = data['YearsCodePro'].apply(categorize_experience)
```
